```python
import numpy as np
from scipy import interp
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.model_selection import StratifiedKFold

%matplotlib inline
pd.set_option('display.max_columns',500)
pd.set_option('display.max_rows',500)
```

```python
df = pd.read_pickle('../data/data.pickle')
```

```python
df.iloc[[0,1,2]]
```

```python
skf = StratifiedKFold(n_splits=4)
skf.get_n_splits(df, df['AFclass'])

print(skf)  
counter = 1
for train_index, test_index in skf.split(df, df['AFclass']):
    split_df = df.iloc[test_index]
    if counter == 4:
        split_df.to_csv('../data/central_data.csv',encoding='utf-8', index=False)
    else:
        split_df.to_csv('../data/hospital' + str(counter) + '_data.csv',encoding='utf-8', index=False)
    counter+=1
    print(split_df.shape)
```
