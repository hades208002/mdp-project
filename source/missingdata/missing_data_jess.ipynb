## 3) Age (no missing)

```python
Age = list(data.Age)
len(set(Age))
```

```python
# There are 49 age groups, let's see the Age Distribution.
age_dict = {}
for age in set(Age):
    age_dict[age] = Age.count(age)

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,1,1)
ax.set_title('Age Distribution', fontsize = 20)
ax.set_xlabel('Age', fontsize = 15)
ax.set_ylabel('Number of Patients', fontsize = 15)
plt.xticks(range(len(age_dict)), list(age_dict.keys()))
plt.bar(range(len(age_dict)), age_dict.values(), align="center")
ax.grid()      
```

```python
# How about the gender distribution within each age group?   {'männlich', 'weiblich'}
age_male = data[data['patsex'] == 'männlich']
age_female = data[data['patsex'] == 'weiblich']
male_age_dict = {}
female_age_dict = {}
for age in Age:
    male_age_dict[age] = len(age_male[age_male['Age']==age])
    female_age_dict[age] = len(age_female[age_female['Age']==age])

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,1,1)
ax.set_title('Age Distribution', fontsize = 20)
ax.set_xlabel('Age', fontsize = 15)
ax.set_ylabel('Number of Patients', fontsize = 15)
plt.xticks(range(len(male_age_dict)), list(male_age_dict.keys()))
plt.bar(range(len(male_age_dict)), list(male_age_dict.values()), align="center", label='male')
plt.bar(range(len(female_age_dict)), list(female_age_dict.values()), bottom=list(male_age_dict.values()), align="center", label='female')
plt.legend()
ax.grid()      
```

```python
# How about also consider the disease class?   {'paroxysmal', 'persistierend (>7 Tage, EKV)'}
age_male_par = data[(data['patsex'] == 'männlich')&(data['AFclass'] == 'paroxysmal')]
age_female_par = data[(data['patsex'] == 'weiblich')&(data['AFclass'] == 'paroxysmal')]
age_male_per = data[(data['patsex'] == 'männlich')&(data['AFclass'] == 'persistierend (>7 Tage, EKV)')]
age_female_per = data[(data['patsex'] == 'weiblich')&(data['AFclass'] == 'persistierend (>7 Tage, EKV)')]
male_par_age_dict = {}
female_par_age_dict = {}
male_per_age_dict = {}
female_per_age_dict = {}
for age in Age:
    male_par_age_dict[age] = len(age_male_par[age_male_par['Age']==age])
    female_par_age_dict[age] = len(age_female_par[age_female_par['Age']==age])
    male_per_age_dict[age] = len(age_male_per[age_male_per['Age']==age])
    female_per_age_dict[age] = len(age_female_per[age_female_per['Age']==age])

m_par = list(male_par_age_dict.values())
f_par = list(female_par_age_dict.values())
m_per = list(male_per_age_dict.values())
f_per = list(female_per_age_dict.values())
    
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,1,1)
ax.set_title('Age Distribution', fontsize = 20)
ax.set_xlabel('Age', fontsize = 15)
ax.set_ylabel('Number of Patients', fontsize = 15)
plt.xticks(range(len(male_par_age_dict)), list(male_par_age_dict.keys()))
plt.bar(range(len(male_par_age_dict)), m_par, align="center", label='male_paroxysmal')
plt.bar(range(len(male_par_age_dict)), m_per, align="center", label='male_persistierend', bottom=m_par, color='blue')

plt.legend()
ax.grid()
```

## 6) Pdur_ 

```python
features = set(data.keys())
Pdur_ = []
for feature in features:
    if feature.find('Pdur_') != -1:
        Pdur_.append(feature)
Pdur_.append('PCdur')
Pdur_.append('AFclass')
```

```python
data_Pdur_ = data[Pdur_]
# replace AFclass 'paroxysmal' -> 0; 'persistierend (>7 Tage, EKV)' -> 1
data_Pdur_['AFclass'].replace('persistierend (>7 Tage, EKV)', 1, inplace=True)
data_Pdur_['AFclass'].replace('paroxysmal', 0, inplace=True)
data_Pdur_.head()
```

```python
# handle missing data with mean value
data_Pdur0 = data_Pdur_[data_Pdur_['AFclass']==0].drop(columns=['AFclass'])
data_Pdur1 = data_Pdur_[data_Pdur_['AFclass']==1].drop(columns=['AFclass'])

means0 = dict(data_Pdur0.mean())
means1 = dict(data_Pdur1.mean())

fig = plt.figure(figsize = (20,10))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('features', fontsize = 15)
ax.set_ylabel('mean value', fontsize = 15)
ax.set_title('Pwave duration analysis', fontsize = 20)

plt.plot(means0.keys(), means0.values(),label='0',color='b')
plt.plot(means1.keys(), means1.values(),label='1',color='r')
ax.legend()
ax.grid()
```

```python
# handle missing data with median value
media0 = dict(data_Pdur0.median())
media1 = dict(data_Pdur1.median())

fig = plt.figure(figsize = (20,10))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('features', fontsize = 15)
ax.set_ylabel('median value', fontsize = 15)
ax.set_title('Pwave duration analysis', fontsize = 20)

plt.plot(media0.keys(), media0.values(),label='0',color='b')
plt.plot(media1.keys(), media1.values(),label='1',color='r')
ax.legend()
ax.grid()
```

```python
# handle missing data with mode value
mode0 = dict(data_Pdur0.mode().mean())
mode1 = dict(data_Pdur1.mode().mean())

fig = plt.figure(figsize = (20,10))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('features', fontsize = 15)
ax.set_ylabel('mode value', fontsize = 15)
ax.set_title('Pwave duration analysis', fontsize = 20)

plt.plot(mode0.keys(), mode0.values(),label='0',color='b')
plt.plot(mode1.keys(), mode1.values(),label='1',color='r')
ax.legend()
ax.grid()
```

### According to the figures, mean value could be used to handling Pdur_ missings.

```python
# check correlation
corr = data_Pdur_.corr()
plt.subplots(figsize=(12,10))
sns.heatmap(corr, square= True, annot = True)
```

For those rows with 'NaN' in every column, we can fill with mean value regard to AFclass.
For those rows with partial missing data, we can predict its value due to the high correlation.

```python
# missing data analysis
miss = {}
for feature in Pdur_:
    if feature != 'AFclass':
        miss[feature] = {}
        missing = set(data[data[feature].isnull()].Soggetti)
        for mis in missing:
            miss[feature][mis] = feature

# plot missings            
fig = plt.figure(figsize = (20,10))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Soggetti', fontsize = 15)
ax.set_ylabel('features', fontsize = 15)
ax.set_title('Pwave duration missing analysis', fontsize = 20)

for feature in Pdur_:
    if feature != 'AFclass':
        plt.scatter(miss[feature].keys(), miss[feature].values())
ax.grid()            
```
